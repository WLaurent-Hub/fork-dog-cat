# Création d'un jeu de donnée pour le modèle GLM

### 1.1 Fusion de chaque année dans une même table

#### 1.1.1 Import des packages de traitement de donnée

```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

#### 1.1.2 Lecture de la couche communes sous format geojson

```python
data = gpd.read_file('../communes.geojson')
```

```python
data.head()
```

#### 1.1.3 Lecture des données géocodées pour chaque année (2017, 2018, 2019, 2020)

```python
data_2017 = pd.read_csv("../../datageocoding/data/2017-geocode.csv", sep=";")
data_2018 = pd.read_csv("../../datageocoding/data/2018-geocode.csv", sep=";")
data_2019 = pd.read_csv("../../datageocoding/data/2019-geocode.csv", sep=";")
data_2020 = pd.read_csv("../../datageocoding/data/2020-geocode.csv", sep=";")
```

```python
data_2017.head()
```

### 1.2 Jointure attributaire des communes geojson et les données 2017

```python
df = data.merge(data_2017, left_on='nom_comm', right_on='VILLE_3', how='left')
```

```python
df
```

```python
df
```

#### 1.2.1 Remplace les valeurs NaN des champs 'CHAT' et 'CHIEN' par 0

```python
df.loc[:, 'CHAT'] = df['CHAT'].fillna(0)
df.loc[:, 'CHIEN'] = df['CHIEN'].fillna(0)
```

```python
df
```

#### 1.2.2 Renomme les champs par année de population

```python
df = df.rename(columns={'CHIEN': 'CHIEN_2017', 'CHAT': 'CHAT_2017'})
```

```python
df
```

### 1.3 Effectue la jointure pour les trois autres année (2018, 2019 et 2020)

#### 1.3.1 Fusion pour les données 2018

```python
df = df.merge(data_2018, left_on='nom_comm', right_on='VILLE_3', how='left')
df.loc[:, 'CHAT'] = df['CHAT'].fillna(0)
df.loc[:, 'CHIEN'] = df['CHIEN'].fillna(0)
df = df.rename(columns={'CHIEN': 'CHIEN_2018', 'CHAT': 'CHAT_2018'})
df = df[[
    'nom_comm', 
    'CHAT_2017', 
    'CHAT_2018', 
    'CHIEN_2017', 
    'CHIEN_2018'
    ]]
```

#### 1.3.2 Fusion pour les données 2019

```python
df = df.merge(data_2019, left_on='nom_comm', right_on='VILLE_3', how='left')
df.loc[:, 'CHAT'] = df['CHAT'].fillna(0)
df.loc[:, 'CHIEN'] = df['CHIEN'].fillna(0)
df = df.rename(columns={'CHIEN': 'CHIEN_2019', 'CHAT': 'CHAT_2019'})
df = df[[
    'nom_comm', 
    'CHAT_2017', 
    'CHAT_2018', 
    'CHAT_2019', 
    'CHIEN_2017', 
    'CHIEN_2018', 
    'CHIEN_2019'
    ]]
```

#### 1.3.3 Fusion pour les données 2020

```python
df = df.merge(data_2020, left_on='nom_comm', right_on='VILLE_3', how='left')
df.loc[:, 'CHAT'] = df['CHAT'].fillna(0)
df.loc[:, 'CHIEN'] = df['CHIEN'].fillna(0)
df = df.rename(columns={'CHIEN': 'CHIEN_2020', 'CHAT': 'CHAT_2020'})
df = df[[
    'nom_comm', 
    'CHAT_2017', 
    'CHAT_2018', 
    'CHAT_2019', 
    'CHAT_2020',
    'CHIEN_2017', 
    'CHIEN_2018', 
    'CHIEN_2019',
    'CHIEN_2020'
    ]]
```

```python
df
```

### 1.4 Interpolation linéaire des valeurs manquantes

#### 1.4.1 Colonne à interpoler

```python
columns_to_interpolate = ['CHAT_2019', 'CHIEN_2019']
```

#### 1.4.2 Interpolation linéaire basé sur les année 2017, 2018 et 2020

```python
for col in columns_to_interpolate:

    for i, row in df.iterrows():
        
        x2 = 2018
        y2 = int(row[f'{col[:-4]}2018'])
        
        x3 = 2020
        y3 = int(row[f'{col[:-4]}2020'])

        x_pred = 2019

        y_pred = int(y2 + (x_pred - x2) * ((y3 - y2) / (x3 - x2)))
        df.at[i, col] = y_pred
```

```python
df
```

### 1.5 Exportation du dataframes au format Excel

```python
df.to_excel("format_data.xlsx", index=False)
```
